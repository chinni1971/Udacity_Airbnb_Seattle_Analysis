```python
def remove_currency_signs(df, col):
    """Remove dollar sign and thousand separator."""
    df[col] = df[col].str.replace('$', '', regex=False)
    df[col] = df[col].str.replace(',', '')
```

```python
def convert_to_bool(df, col):
    """Convert object columns to boolean."""
    df[col] = df[col].replace({'f': 0, 't': 1})
    df[col] = df[col].astype(bool)
```

### Q3 Do high performing hosts charge more?
Now that we know that there is a significant difference between pricing for different neighborhoods, I wanted to find out whether hosts in these areas with high performance scores tend to charge higher prices.

```python
def price_diff(price_col, score_col, cat_col):
    """
    Compare prices based on a host's review score, separated by a categorical column.
    
    ARGUMENTS:
    price_col - price variable that will be compared
    score_col - scoring metric
    cat_col - categorical variable to group the dataframe
    
    RETURN VALUE:
    df - dataframe
    """
    
    # Dataset split exactly in half, alternatives: 75th percentile or the mean
    threshold = df_listings[score_col].quantile(0.5)
    # High performers belong to the better 50%
    upper_df = df_listings[df_listings[score_col] >= threshold].groupby(cat_col)[price_col].mean().sort_values(ascending=False).to_frame().reset_index()
    upper_df = upper_df.rename(columns={price_col:'{} High Performers (USD)'.format(price_col.capitalize())})
    # Low performers belong to the lower 50%
    lower_df = df_listings[df_listings[score_col] < threshold].groupby(cat_col)[price_col].mean().sort_values(ascending=False).to_frame().reset_index()
    lower_df = lower_df.rename(columns={price_col:'{} Low Performers (USD)'.format(price_col.capitalize())})
    
    df = upper_df.merge(lower_df, on=cat_col)
    df['Difference (USD)'] = df[df.columns[1]] - df[df.columns[2]]
    
    return df 
    
```

```python
df_sub = price_diff('price', 'review_scores_rating', 'neighbourhood_group_cleansed')
df_sub.rename(columns={'neighbourhood_group_cleansed':'Neighborhood'}).style.bar(subset=['Difference (USD)'], align='mid', color=['#d65f5f', '#5fba7d'])
```

**Results:** In general, hosts who perform well according to their rating also charge more if they are located in a higher priced neighbourhood to begin with. Example: Magnolia represents the most exclusive area in Seattle in terms of prices per night. Hosts who belong to the better half on average charge USD 71 more than the other 50%. For neighbourhoods which belong to the cheaper locations in Seattle this trend inverts. In Delridge high performers on average charge almost USD 9 less. This trend might be explained by the importance of pricing on a visitors booking decision. If I already search for cheaper stays, the host's performance might be less relevant than the price per night. On the contrary, if I am interested in exclusive areas, I might be inclined to pay even more for a listing with high ratings.

## Modeling
In the next section I will create two linear regression models to determine which features influence pricing and positive reviews the most.

```python
# Create a subset from the listings dataframe
df_model = df_listings.drop(['id', 'host_id', 'description', 'amenities', 'weekly_price', 'monthly_price',
                             'host_since', 'first_review', 'last_review'], axis=1)
```

```python
# Combine property types which appear less than 10 times into the Other category
# see section above (influence house boat on results)
tmp = (df_model.property_type.value_counts() < 10).to_frame().reset_index()
prop_types = tmp[tmp.property_type]['index'].values
```

```python
df_model.loc[df_model[df_model.property_type.isin(prop_types)].index, 'property_type'] = 'Other'
```

```python
def get_dummies_df(df, cat_cols):
    """
    Create dummy variables for categorical columns.
    
    ARGUMENTS:
    df - dataframe to be transformed
    cat_cols - categorical columns
    
    RETURN VALUE:
    df - transformed dataframe
    """

    for col in cat_cols:
        try:
            df = pd.concat([df.drop(col, axis=1), pd.get_dummies(df[col], prefix=col, prefix_sep='_', drop_first=True)], axis=1)
        except:
            continue
    
    # Normalize column names
    df.columns = [col.replace(' ', '_').lower() for col in df.columns]
    df.columns = [col.replace('-', '').lower() for col in df.columns]
    
    try:
        df.rename(columns={'property_type_bed_&_breakfast': 'property_type_bed_breakfast'}, inplace=True)
        df.rename(columns={'property_type_camper/rv': 'property_type_camper_rv'}, inplace=True)
    except:
        pass
    
    return df
```

```python
def linear_regression_model(df, target_col):
    """
    Create a linear regression model fitted on a given training dataset.
    
    ARGUMENTS:
    df - dataframe containing all data
    target_col - target variable name
    
    RETURN VALUES:
    X_train - independent variables in training dataset
    lm_model - trained regression model
    test_score - r2 score on the test data
    train_score - r2 score on the train data
    """
    
    X = df.drop(target_col, axis=1)
    Y = df[target_col]
    
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=3)
    
    lm_model = LinearRegression(normalize=True)
    lm_model.fit(X_train, Y_train)
    
    Y_pred_test = lm_model.predict(X_test)
    Y_pred_train = lm_model.predict(X_train)
    
    test_score = r2_score(Y_test, Y_pred_test)
    train_score = r2_score(Y_train, Y_pred_train)
    
    return X_train, lm_model, test_score, train_score
```

```python
def get_coefs_df(X_train, lm_model):
    """
    Create dataframe displaying the correlation coefficients for each variable.
    
    ARGUMENTS:
    X_train - independent variables from training dataset
    lm_model - trained linear regression model
    
    RETURN VALUE:
    coefs_df - dataframe with correlation coefficients in descending order
    """
    
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    coefs_df['abs_coefs'] = np.abs(lm_model.coef_)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    
    return coefs_df
```

## Evaluation and Deployment

The evaluation metric used for both linear regression models is the r2 score. It measures the proportion of variance in the dependent variable which can be explained by the independent variables. In the first scenario (target: price) over 60% of the fitted data can be explained by the linear model. For the second scenario (target: rating) more than 70% of the test dataset can be explained by the model.

**Interpreting the coefficients**

The coefficients for the independent variables indicate how much they influence the target variable. They should be interpreted in relation to one another. It is important to point out that influence can go in both directions - price increase and decrease.

* *bedrooms* and *room_type_shared_room* are both a measure of the number of guests (see appendix, figures A and B). With more bedrooms the number of guests usually increases and thus the price charged per night. Vice versa a shared room means I am staying in a room with other guests (like in a hostel). The price charged per night decreases.
* The *property_type* was already highlighted during the descriptive statistics section. Loft and Bed & Breakfast only appear 31 and 25 times respectively in our dataset. Further analysis would be required to understand the impact of a specific property type on pricing.
* It is the location (*neighbourhood_group_cleansed*) which has a large impact on pricing for an Airbnb listing. Magnolia, Queen Anne, Dowtown and Cascade are the most expensive areas in Seattle and rank among the top ten coefficients. The fact that a host offers a listing in these locations alone increases the value per night. Whether or not a host has a strict cancellation policy, is verified, or communicates frequently is much less important. 

It all comes down to location, location, location.
